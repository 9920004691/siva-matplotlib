```python
# Import pandas and read the csv file
import pandas as pd
data = pd.read_csv("/content/cri t20.csv")
```

```python
# Display the first ten rows
data.head()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
x = data.Player
y = data.Inns
my_circle = plt.Circle( (0,0), 0.4, color='white')
plt.pie(y, labels=x, colors=['b','g','r','c','m','y','k','w','r','c','m','y','k','b','g','r','c','m','y'], wedgeprops={'linewidth':6,'edgecolor':'white'})       
plt.title("Cut Pie")
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = data.Player
y = data.Ave
my_circle = plt.Circle( (0,0), 0.6, color='white')
plt.pie(y, labels=x, colors=['b','g','r','c','m','y','k','c','r','c','m','y','k','b','g','r','c','m','y'])
plt.title("Donut Chart")
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
# importing libraries
import matplotlib.pyplot as plt
import seaborn
y = data.Player
x = data.Ave
palette_color = seaborn.color_palette('bright')
plt.pie(x, labels=y, colors=palette_color, radius = 2,autopct='%.0f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn
y = data.Player
x = data.Ave
palette_color = seaborn.color_palette('bright')
my_circle = plt.Circle( (0,0), 1, color='white')
plt.pie(x, labels=y, colors=palette_color,radius = 2,shadow = 'True',autopct='%.0f%%')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```
